```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = "/content/drive/MyDrive/Au_nanoparticle_dataset.csv"
df = pd.read_csv(path)

new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

print(new_df.head(20))

print(new_df.describe())

new_df.hist(layout=(1, 4), figsize=(15, 3))
plt.tight_layout()
plt.show()

sns.pairplot(new_df)
plt.show()

g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot, cmap='Blues')
plt.show()
```
